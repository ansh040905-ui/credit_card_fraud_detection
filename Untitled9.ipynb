```python
# Core libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time, copy, warnings

# Classical models and metrics
from sklearn import linear_model, svm
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score
from sklearn.datasets import make_moons # For a simple, runnable dataset

# Quantum libraries (with corrected imports)
from qiskit.circuit.library import PauliFeatureMap, ZFeatureMap, ZZFeatureMap
from qiskit_machine_learning.circuit.library import RawFeatureVector
from qiskit.circuit.library import RealAmplitudes
from qiskit_algorithms.optimizers import COBYLA
from qiskit.primitives import Sampler
from qiskit_machine_learning.algorithms.classifiers import VQC
from qiskit_aer.primitives import Sampler as AerSampler # Use an alias to avoid conflict
from IPython.display import clear_output

# Define a callback function to show progress during training
def callback(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.xlabel("Iteration")
    plt.ylabel("Objective function value")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()

# --- 1. DATA LOADING AND PREPROCESSING ---
# Using a simple, built-in dataset to avoid FileNotFoundError and ParserError
print("Loading and preprocessing data...")
X, y = make_moons(n_samples=100, noise=0.05, random_state=42)

# Scale the features
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Data loading and preprocessing complete.\n")

# --- 2. QUANTUM KERNEL AND MODEL SETUP ---
# Define the feature map and ansatz
n_features = X_train.shape[1]
fm = ZZFeatureMap(feature_dimension=n_features, reps=1, entanglement='linear')
pqc = RealAmplitudes(num_qubits=n_features, reps=1)

# Define the Sampler primitive and the VQC optimizer
sampler = Sampler()
optimizer = COBYLA(maxiter=50)
objective_func_vals = []

# Instantiate the VQC classifier with corrected parameters
vqc = VQC(
    sampler=sampler,
    feature_map=fm,
    ansatz=pqc,
    optimizer=optimizer,
    callback=callback
)

# --- 3. MODEL TRAINING AND EVALUATION ---
print("Training VQC model...")
vqc.fit(X_train, y_train)

# Make predictions on the training and test sets
print("\nMaking predictions...")
y_train_pred = vqc.predict(X_train)
y_test_pred = vqc.predict(X_test)

# Print classification reports and confusion matrices
print("\n--- Training Set Results ---")
print(classification_report(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred), "= confusion matrix for train set \n")

print("--- Test Set Results ---")
print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred), "= confusion matrix for test set \n ")
print(f"{round(roc_auc_score(y_test, y_test_pred), 2)} = roc_auc_score for test set \n")

print("Code execution complete.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Qiskit imports
from qiskit import QuantumCircuit
from qiskit.circuit import ParameterVector
from qiskit.circuit.library import RealAmplitudes
from qiskit_algorithms.optimizers import COBYLA
from qiskit.primitives import Sampler
from qiskit_machine_learning.algorithms.classifiers import VQC

# Metrics
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score

# --- Define variables ---
num_features = 2   # number of features
num_qubits = 2     # number of qubits
training_size = 50
testing_size = 20

# Generate toy dataset (binary classification)
X_train = np.random.rand(training_size, num_features)
y_train = np.random.randint(0, 2, training_size)

X_test = np.random.rand(testing_size, num_features)
y_test = np.random.randint(0, 2, testing_size)

# --- FEATURE MAP (Angle Encoding using RY gates) ---
fm = QuantumCircuit(num_qubits)
# Correct way: use ParameterVector for angles
features = ParameterVector('x', num_features)
for i in range(num_features):
    fm.ry(features[i], i)

# Ansatz (variational circuit)
pqc = RealAmplitudes(num_qubits=num_qubits, reps=1)

# Optimizer & Sampler
optimizer = COBYLA(maxiter=50)
sampler = Sampler()

# --- Instantiate and Train the VQC ---
vqc = VQC(
    sampler=sampler,
    feature_map=fm,
    ansatz=pqc,
    optimizer=optimizer
)

print("⏳ Training VQC model...")
vqc.fit(X_train, y_train)
print("✅ Training complete.\n")

# --- Predictions ---
print("🔮 Making predictions...")
y_train_pred = vqc.predict(X_train)
y_test_pred = vqc.predict(X_test)

# --- Results ---
print("\n=== Training Set Results ===")
print("Accuracy:", round(accuracy_score(y_train, y_train_pred), 3))
print(classification_report(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred), "= confusion matrix (train set)\n")

print("=== Test Set Results ===")
print("Accuracy:", round(accuracy_score(y_test, y_test_pred), 3))
print(classification_report(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred), "= confusion matrix (test set)\n")
print("ROC-AUC Score:", round(roc_auc_score(y_test, y_test_pred), 3))

print("\n🎉 Code execution complete.")
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')

import numpy as np
from sklearn import linear_model, svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- ADDED: Define and prepare the data ---
# This part is necessary to make the code runnable
# We'll use dummy data for demonstration
X = np.random.rand(200, 20)
y = np.random.randint(0, 2, 200)

# Scale the data for better performance with some models
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
# ---------------------------------------------------

# --- Define a function to train & evaluate ---
def train_test(ML_model, model_name="Model"):
    print("="*70)
    print(f"▶️ Training {model_name}\n")

    # Fit model
    ML_model.fit(X_train, y_train)

    # Predictions
    y_train_pred = ML_model.predict(X_train)
    y_test_pred = ML_model.predict(X_test)

    # --- Train results ---
    print("📊 Training Results")
    print("Accuracy:", round(accuracy_score(y_train, y_train_pred), 3))
    print(classification_report(y_train, y_train_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_train, y_train_pred), "\n")

    # --- Test results ---
    print("📊 Test Results")
    print("Accuracy:", round(accuracy_score(y_test, y_test_pred), 3))
    print(classification_report(y_test, y_test_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
    print("ROC-AUC Score:", round(roc_auc_score(y_test, y_test_pred), 3))
    print("="*70, "\n")

# ----------------------------------------------------------------------------------

# Try different models
train_test(linear_model.LogisticRegression(), "Logistic Regression")
train_test(MLPClassifier(max_iter=500), "Neural Network (MLP)")
train_test(svm.SVC(probability=True), "Support Vector Machine")
train_test(RandomForestClassifier(), "Random Forest")
```
